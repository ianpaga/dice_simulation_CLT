# Average of dice rolls vs probability density

```python
import numpy as np
import matplotlib.pyplot as plt

def roll_dice(num_rolls, num_dice):

    """
    Simulates rolling a specified number of dice multiple times and computes the average of the outcomes for each roll.

    Parameters:
    - num_rolls (int): Number of times to roll the dice.
    - num_dice (int): Number of dice to roll in each trial.

    Returns:
    - averages (numpy.ndarray): Array containing the average outcome of each trial.
    """
    
    rolls = np.random.randint(1, 7, (num_rolls, num_dice))
    averages = np.mean(rolls, axis=1)
    return averages

def plot_histogram(ax, data, num_bins, title, show_xlabel=False, show_ylabel=False):

    """
    Plots a histogram of the given data.

    Parameters:
    - ax (matplotlib.axes.Axes): The axes object to draw the plot onto.
    - data (numpy.ndarray): Array of data to plot.
    - num_bins (int): Number of bins for the histogram.
    - title (str): Title of the plot.
    - show_xlabel (bool, optional): Whether to show the x-label. Defaults to False.
    - show_ylabel (bool, optional): Whether to show the y-label. Defaults to False.
    """
    
    ax.hist(data, bins=num_bins, edgecolor='black', density=True)
    ax.set_title(title)
    if show_xlabel:
        ax.set_xlabel('Average of Dice Rolls')
    if show_ylabel:
        ax.set_ylabel('Probability Density')

num_rolls = 10000
dice_configs = [1, 2, 3, 5, 10, 50, 100, 1000, 10000]

fig, axs = plt.subplots(3, 3, figsize=(8, 6))
fig.suptitle('Fair dice')
axs = axs.flatten()

for i, num_dice in enumerate(dice_configs):
    show_xlabel = (i >= 6)  # Show x-labels for the bottom row
    show_ylabel = (i % 3 == 0)  # Show y-labels for the first column
    averages = roll_dice(num_rolls, num_dice)
    plot_histogram(axs[i], averages, num_bins=30, title=f'{num_dice} Dice', show_xlabel=show_xlabel, show_ylabel=show_ylabel)


plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def roll_dice(num_rolls, num_dice):

    """
    Simulates rolling a specified number of dice multiple times and computes the average of the outcomes for each roll.

    Parameters:
    - num_rolls (int): Number of times to roll the dice.
    - num_dice (int): Number of dice to roll in each trial.

    Returns:
    - averages (numpy.ndarray): Array containing the average outcome of each trial.
    """
    
    # Define the probabilities for each outcome of a fair 6-sided dice
    # Example: for an unfair dice, you could change these probabilities accordingly
    probabilities = [1/12, 1/12, 1/6, 1/6, 1/3, 1/6]
    
    # Rolls for each die
    rolls = np.random.choice(np.arange(1, 7), size=(num_rolls, num_dice), p=probabilities)
    
    # Average of the rolls for each trial
    averages = np.mean(rolls, axis=1)
    return averages

def plot_histogram(ax, data, num_bins, title, show_xlabel=False, show_ylabel=False):

    """
    Plots a histogram of the given data.

    Parameters:
    - ax (matplotlib.axes.Axes): The axes object to draw the plot onto.
    - data (numpy.ndarray): Array of data to plot.
    - num_bins (int): Number of bins for the histogram.
    - title (str): Title of the plot.
    - show_xlabel (bool, optional): Whether to show the x-label. Defaults to False.
    - show_ylabel (bool, optional): Whether to show the y-label. Defaults to False.
    """
    
    ax.hist(data, bins=num_bins, edgecolor='black', density=True)
    ax.set_title(title)
    if show_xlabel:
        ax.set_xlabel('Average of Dice Rolls')
    if show_ylabel:
        ax.set_ylabel('Probability Density')

num_rolls = 10000
dice_configs = [1, 2, 3, 5, 10, 50, 100, 1000, 10000]

fig, axs = plt.subplots(3, 3, figsize=(8, 6))
fig.suptitle('Non-fair dice')
axs = axs.flatten()

for i, num_dice in enumerate(dice_configs):
    show_xlabel = (i >= 6)      # Show x-labels for the bottom row
    show_ylabel = (i % 3 == 0)  # Show y-labels for the first column
    averages = roll_dice(num_rolls, num_dice)
    plot_histogram(axs[i], averages, num_bins=30, title=f'{num_dice} Dice', show_xlabel=show_xlabel, show_ylabel=show_ylabel)

plt.tight_layout()
#plt.savefig('nonfair_dice_prob_density.png', dpi=500)
plt.show()
```

**Central Limit Theorem (CLT)**

The Central Limit Theorem states that the distribution of the `sum` (or `average`) of a large number of independent, identically distributed variables approaches a `normal` (Gaussian) distribution, regardless of the original distribution of the variables. In this code:

Each dice roll represents a random variable.
As the number of dice increases, the distribution of the sum of the dice rolls becomes more Gaussian, illustrating the CLT.
The histograms generated in the code show this progression towards a normal distribution as the number of dice increases.
